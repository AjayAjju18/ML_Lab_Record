```python
# 19. Prepare X and y

X = df.drop('Species', axis=1) #Independent All features Except Species
y = df['Species']              #Dependent
```

```python
# 20. Train-test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 21–23. Import and instantiate SVC

from sklearn.svm import SVC
model = SVC(C=1)


## help(SVC)
```

```python
# 24. Fit model

model.fit(X_train, y_train)
```

```python
# 25. Predict

y_pred = model.predict(X_test)
```

```python
# 26. Confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
# 27. Other metrics

from sklearn.metrics import accuracy_score, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# 28. Classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# 29–30. K-Fold Cross Validation

from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, y, cv=5)
print("CV Scores:", scores)
print("Mean CV Accuracy:", scores.mean())
```

```python
# 31–36. Grid Search for best C value

from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000]}
grid = GridSearchCV(SVC(), param_grid, cv=5)
grid.fit(X, y)

print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

```python
# 34. Results as DataFrame

results_df = pd.DataFrame(grid.cv_results_)
print(results_df[['param_C', 'mean_test_score']])
```

```python
# 35. Plot C vs accuracy

plt.plot(results_df['param_C'], results_df['mean_test_score'], marker='o')
plt.title("C vs Mean Test Score")
plt.xlabel("C Value")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```
